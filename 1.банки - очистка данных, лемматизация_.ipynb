### Лемматизация  <a name="шаг-2-4"></a>

Столбец purpose заполнен неоднородным текстом с различными целями клиентов. Проанализировать их быстро и качественно без помощи лемматизации очень сложно. Также используем контейнер Counter для подсчёта встречаемости каждой категории:

```python
from pymystem3 import Mystem
m = Mystem()

def lemmas(purpose):
    lem_list = ' '.join(m.lemmatize(purpose))
    return lem_list

df['lemma_purpose'] = df['purpose'].apply(lemmas)

```

```python
from collections import Counter
Counter(df['lemma_purpose'])
```
